```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt


img = Image.open('timidez.jpg')


filtros = [
    ('Original', None),
    ('BLUR', ImageFilter.BLUR),
    ('CONTOUR', ImageFilter.CONTOUR),
    ('DETAIL', ImageFilter.DETAIL),
    ('EDGE_ENHANCE', ImageFilter.EDGE_ENHANCE),
    ('EMBOSS', ImageFilter.EMBOSS),
    ('FIND_EDGES', ImageFilter.FIND_EDGES),
    ('SHARPEN', ImageFilter.SHARPEN),
    ('GaussianBlur', ImageFilter.GaussianBlur(radius=3)),
    ('ModeFilter', ImageFilter.ModeFilter(size=3))
]


imagenes_filtradas = []
for nombre, filtro in filtros:
    if filtro is None:
        imagen = img.copy()
    else:
        imagen = img.filter(filtro)
    imagenes_filtradas.append((nombre, imagen))


columnas = 3
filas = (len(imagenes_filtradas) + columnas - 1) // columnas
fig, axs = plt.subplots(filas, columnas, figsize=(12, 4 * filas))


axs = axs.flatten()


for i, (nombre, imagen) in enumerate(imagenes_filtradas):
    axs[i].imshow(imagen)
    axs[i].set_title(nombre)
    axs[i].axis('off')

# Desactivar subplots vacíos
for j in range(i + 1, len(axs)):
    axs[j].axis('off')

plt.tight_layout()
plt.show()
```
